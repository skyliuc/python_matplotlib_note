# 简单的画散点图

```python
#导入必要的模块
import numpy as np
import matplotlib.pyplot as plt
#产生测试数据
x = np.arange(1,10)
y = x
fig = plt.figure()
ax1 = fig.add_subplot(111)
#设置标题
ax1.set_title('Scatter Plot')
#设置X轴标签
plt.xlabel('X')
#设置Y轴标签
plt.ylabel('Y')
#画散点图
ax1.scatter(x,y,c = 'c',marker = '1')
#设置图标
plt.legend('x1')
#显示所画的图
plt.show()
```

#  用类别变量来画图

```python
#就是说你的横坐标可以是类别，这个挺方便的
names=['group_a','group_b','group_c']
values=[1,10,100]
plt.figure(1,figsize=(9,3))
plt.subplot(131)
plt.bar(names,values)
#柱状图
plt.subplot(132)
plt.plot(names,values)
plt.subplot(133)
plt.scatter(names,values)
plt.suptitle('Categorical Plotting')
plt.show()
```

# 画出自定义的函数图像

```python
#画出函数图像
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.figure(1)
plt.subplot(211)
plt.plot(t1,f(t1),'co',t2,f(t2),'m')
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```

# 给画出的图像注释，顺便学学其它操作

```python
#画出IQ图
mu,sigma=100,15
x=mu+sigma*np.random.randn(10000)
n,bins,pathes=plt.hist(x,50,density=1,facecolor='c',alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60,0.025,r'$\mu=100,\ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

# 用箭头指示

```python
#画出箭头来注释
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='red', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```

# 用subplot展示多个图

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1961)
data = np.random.randn(2, 100)

fig, axs = plt.subplots(2, 2, figsize=(5, 5))
#figsize图像的大小
axs[0, 0].hist(data[0])
#左上
axs[1, 0].scatter(data[0], data[1])
#右上
axs[0, 1].plot(data[0], data[1])
#左下
axs[1, 1].hist2d(data[0], data[1])
#右下
plt.show()
```

```python
#首先是老套路
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#不解释的素质三连续
```

# 数据　data

```python
from matplotlib.ticker import FuncFormatter

data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
print('The value of data in group:\n' )
print(group_data)
print("##########################################################################################################\n")
print('The names in group:\n')
print(group_names)
print("##########################################################################################################\n")
print('The mean of goup value:\n')
print(group_mean)
```

给图加上注释

改变图像的大小可以改变figsize的参数

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
```

```python
高端玩法，改变坐标轴，之前是用数字，这次用字符串
```

```python
def currency(x, pos):
    """The two args are the value and tick position"""
    if x >= 1e6:
        s = '${:1.1f}M'.format(x*1e-6)
    else:
        s = '${:1.0f}K'.format(x*1e-3)
    return s

formatter = FuncFormatter(currency)

fig, ax = plt.subplots(figsize=(6, 8))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.xaxis.set_major_formatter(formatter)
```

增加带有更多信息的注释

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

# Add a vertical line, here we set the style in the function call
ax.axvline(group_mean, ls='--', color='r')

# Annotate new companies
for group in [3, 5, 8]:
    ax.text(145000, group, "New Company", fontsize=10,
            verticalalignment="center")

# Now we'll move our title up since it's getting a little cramped
ax.title.set(y=1.05)

ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.xaxis.set_major_formatter(formatter)
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3])
fig.subplots_adjust(right=.1)

plt.show()
```
